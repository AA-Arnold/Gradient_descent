#### In this notebook i'll be writing a gradient descent algorithm using different cost function given a cost function we would find the derivatives which will tell us about the slope of the function which we will plot as well as visualizing how the algorithm descend with each iteration until it gets to the minimal cost we'll observer different learning rate and and how the converge

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Generating data between (-2 - 3)

```python
x_1 = np.linspace(start=-2, stop=3, num=100)
x_1
```

## A simple cost function
## $$f(x)=x^4+x+2$$


```python
def f(x):
    return x**4 + x + 2

f(x_1)
```

## Graphing cost function of - $f(x)=x^4+x+2$

```python
plt.title("Cost Function for f(x)")
plt.xlim([-3, 3])
plt.ylim(1, 30)
plt.xlabel('X', fontsize = 16)
plt.ylabel('f(x)' , fontsize = 16)
plt.plot(x_1, f(x_1))
plt.show()
```

## Using the power rule we find the slope of the function

## Formular to apply the power rule
## $$\frac{d}{dx}  x^{n} = n \cdot x^n-1$$

$$f(x)=x^4+x+2$$
## $$df(x)=4x^3+x$$

```python
def df(x):
    return 4*(x**3) + x
```

## Plot function and derivates side by side

```python
plt.figure(figsize=[15 , 5])

#1 chart : cost function 
plt.subplot(1, 2, 1) #allows ploting in colum 1
plt.title("Cost Function for f(x)")
plt.xlim([-3, 3])
plt.ylim(1, 30)
plt.xlabel('X', fontsize = 16)
plt.ylabel('f(x)' , fontsize = 16)
plt.plot(x_1, f(x_1))


plt.plot(x_1, f(x_1), color='blue', linewidth=3)

#2 chart : Derivative
plt.subplot(1, 2, 2)
plt.title('Derivative')
plt.xlabel('X', fontsize = 16)
plt.ylabel('df(x)' , fontsize = 16)
plt.grid()

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5)
plt.show()
```

### Implementing gradient descent algorithm with initial guess  of 3 and learning rate of 0.01

```python
# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.01
precision = 0.0001

cost_list = [new_x]
slope_list = [df(new_x)]

for n in range(1000):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient
    
    step_size = abs(new_x - previous_x)
    
    cost_list.append(new_x)
    slope_list.append(df(new_x))

    if step_size < precision:
        print('Loop ran this many times ', n)
        break
     
    
print('Local minimum occurs at ', new_x)
print('slope of df(x) value at this point ', df(new_x))
print('f(x) value or cost at this point is ', f(new_x))
```

### using a scatter plot to visualize the step between each iteration

```python
plt.figure(figsize=[15 , 5])

#1 chart : cost function 
plt.subplot(1, 2, 1) 
plt.title("Cost Function for f(x)")
plt.xlim([-3, 3])
plt.ylim(1, 30)
plt.xlabel('X', fontsize = 16)
plt.ylabel('f(x)' , fontsize = 16)
plt.plot(x_1, f(x_1),color='blue', linewidth=3)
plt.scatter(cost_list, f(np.array(cost_list)), color='red', s = 100, alpha =0.6 )


#2 chart : Derivative
plt.subplot(1, 2, 2)
plt.title('Derivative')
plt.xlabel('X', fontsize = 16)
plt.ylabel('df(x)' , fontsize = 16)
plt.grid()

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5)
plt.scatter(cost_list, slope_list, color='red', s = 100, alpha =0.6 )

plt.show()
```

we can observe that at the initial guess of 3 it takes a very large step on first iteration because of the steepness of the slope at that point before steps start to reduce untill it reaches the local minimal

## Graphing cost function of - $g(x) = 2x^2+3x^3-2-0$

```python
def g(x):
    return 2*(x**2)+3*(x**3)-2-0
def dg(x):
    return 4*x+9*(x**2)-2-0
```

```python
# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.015
precision = 0.001

cost_list = [new_x]
slope_list = [dg(new_x)]

for n in range(1000):
    previous_x = new_x
    gradient = dg(previous_x)
    new_x = previous_x - step_multiplier * gradient
    
    cost_list.append(new_x)
    slope_list.append(dg(new_x))

    if step_size < precision:
        print('Loop ran this many times ', n)
        break
     
    
print('Local minimum occurs at ', new_x)
print('slope of dg(x) value at this point ', dg(new_x))
print('g(x) value or cost at this point is ', g(new_x))
```

```python
plt.figure(figsize=[15 , 5])

#1 chart : cost function 
plt.subplot(1, 2, 1) 
plt.title("Cost Function for g(x)")
plt.xlabel('X', fontsize = 16)
plt.ylabel('f(x)' , fontsize = 16)
plt.plot(x_1, g(x_1),color='blue', linewidth=3)
plt.scatter(cost_list, g(np.array(cost_list)), color='red', s = 100, alpha =0.6 )


#2 chart : Derivative
plt.subplot(1, 2, 2)
plt.title('Derivative')
plt.xlabel('X', fontsize = 16)
plt.ylabel('dg(x)' , fontsize = 16)
plt.grid()

plt.plot(x_1, dg(x_1), color='skyblue', linewidth=5)
plt.scatter(cost_list, slope_list, color='red', s = 100, alpha =0.6 )

plt.show()
```

```python
def gradient_descent(derivative_function, random_guess, learning_rate, precision):
    new_x = random_guess
    cost_list = [new_x]
    slope_list = [derivative_function(new_x)]

    for n in range(1000):
        previous_x = new_x
        gradient = derivative_function(previous_x)
        new_x = previous_x - learning_rate * gradient

        cost_list.append(new_x)
        slope_list.append(derivative_function(new_x))

        if step_size < precision:
            print('Loop ran this many times ', n)
            break
            
        return new_x,cost_list,slope_list
```
